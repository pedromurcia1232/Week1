```python
#Codigo Original
import numpy as np
def step(z):
    return 1 if z >= 0 else 0
def neurona(x1, x2, w1, w2, b):
    z = x1*w1 + x2*w2 + b
    y = step(z)
    return z, y
    # Valores base
w1, w2, b = 1, 1, -1.5 # intenta parecerse a una compuerta AND
casos = [(0,0) , (0,1) , (1,0) , (1,1)]
for x1, x2 in casos:
      z, y = neurona(x1, x2, w1, w2, b)
      print(f"Entrada=({x1}, {x2}) z={z:.1f} salida={y}")
```

1. ¿Con qué valor de b la neurona se comporta más parecido a AND?
La neurona se comporta más parecido a una compuerta AND cuando el bias es b = -2.0, ya que solo produce salida 1 cuando ambas entradas son (1,1) y en los demás casos la salida es 0.

2. ¿Qué efecto tiene bajar o subir el bias en la salida?
Al hacer el bias más negativo, la neurona se vuelve más estricta y requiere una mayor suma de las entradas para activarse. Al aumentar el bias (hacerlo menos negativo), la neurona se activa con mayor facilidad, incluso cuando solo una entrada es 1.
